## **Question 1**

**A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from bot units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.**

We are going to conduct 2 tailed t-test on 2 independent samples with numerical data



*   Null Hypothesis:- Unit A = Unit B
*   Alternate Hypothesis: - Unit A != Unit B

significance 5%

Alpha = 0.05

as it is two-tailed test



*   alpha/2 = 0.025



```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data = pd.read_csv('Cutlets.csv')
data.head()
```

```python
data.describe()
```

```python
unit_A = pd.Series(data.iloc[:,0])
unit_A
```

```python
unit_B = pd.Series(data.iloc[:,1])
unit_B
```

```python
# 2-sample 2-tail ttest:   stats.ttest_ind(array1,array2)
# ind -> independent samples

p_value = stats.ttest_ind(unit_A, unit_B)
p_value
```

Compare p_value with Significance level
If p_value is not equal to significance level then we failed to reject Null Hypothesis because of lack of evidence.

If p_value is equal to significance then we reject Null Hypothesis

as p_value is greater than 0.5 i.e 0.025
Hence we fail to reject Null Hypothesis and there is no significant difference between the two samples.
